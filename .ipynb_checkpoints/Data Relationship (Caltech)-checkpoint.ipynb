# Data relationships between day of the week and month of the year with kWhDelivered (Caltech)

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import pytz
from datetime import datetime

# Load the JSON data
with open('acndata_sessions.json', 'r') as file:
    data = json.load(file)

# Extract the items
items = data['_items']

# Create a DataFrame
df = pd.DataFrame(items)

```

```python
# Relationships between day of the year
```

```python
# Define the timezone for the connectionTime
timezone = pytz.timezone('UTC')

# Parse the connectionTime with explicit timezone handling
df['connectionTime'] = pd.to_datetime(df['connectionTime']).dt.tz_localize('UTC').dt.tz_convert('America/Los_angeles')
df['dayOfWeek'] = df['connectionTime'].dt.day_name()
df['month'] = df['connectionTime'].dt.strftime('%B')

# Group by day of the week and sum the kWhDelivered
grouped_by_day = df.groupby('dayOfWeek')['kWhDelivered'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])


# Plot the data for day of the week
plt.figure(figsize=(10, 6))
grouped_by_day.plot(kind='bar', color='skyblue')
plt.title('Total kWh Delivered by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total kWh Delivered')
plt.xticks(rotation=45)
plt.grid(axis='y')
```

```python
# Relationship based on month of the year
```

```python
# Group by month and sum the kWhDelivered
grouped_by_month = df.groupby('month')['kWhDelivered'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'
])

# Plot the data for month of the year
plt.figure(figsize=(10, 6))
grouped_by_month.plot(kind='bar', color='lightgreen')
plt.title('Total kWh Delivered by Month of the Year')
plt.xlabel('Month')
plt.ylabel('Total kWh Delivered')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```
